```python

# Analyze.py
# Original by Annika Muehlbradt
# Modified by Chris Wilcox
# 11/13/2012

# Imports
import dpkt
import os
import socket
import string
import sys

# Main program
if __name__ == "__main__":

    # Packet Counters
    counter=0.0
    ipcounter=0.0
    nonipcounter=0.0    
    tcpcounter=0.0
    udpcounter=0.0
    arpcounter=0.0
    httpcounter=0.0
    httpscounter=0.0
    smtpcounter=0.0
    dhcpcounter=0.0
    ftpcounter=0.0
    sshcounter=0.0
    ntpcounter=0.0
    telnetcounter=0.0
    whoiscounter=0.0
    rsynccounter=0.0
    icmpcounter=0.0
    ipv6counter=0.0

    # Subnet Dictionary
    subnets = {}

    # Open file
    filename="Laksh.pcap"
    if os.path.isfile(filename):
        print ("Present: ",filename)
        trace=filename
    else:
        print ("Absent: ",filename)
        sys.stderr.write("Cannot open file for reading\n")
        sys.exit(-1)

    # Packet processing loop
    for ts,pkt in dpkt.pcap.Reader(open(filename,'rb')):
        counter+=1
        
        # Parse ethernet packet
        eth=dpkt.ethernet.Ethernet(pkt)
        ip=eth.data       
                
        #check if IP packet or non-ip packet
        if eth.type == dpkt.ethernet.ETH_TYPE_IP or eth.type == dpkt.ethernet.ETH_TYPE_IP6:
            ipcounter = ipcounter + 1
        else:
            nonipcounter = nonipcounter + 1    

        # IPV6 packets
        if eth.type==dpkt.ethernet.ETH_TYPE_IP6: 
            ipv6counter+=1     

        # ARP packet
        elif eth.type==dpkt.ethernet.ETH_TYPE_ARP:
            arpcounter+=1

        # ICMP packets
        elif ip.p==dpkt.ip.IP_PROTO_ICMP:
            icmpcounter+=1
            
        # IPV4 packets
        elif eth.type==dpkt.ethernet.ETH_TYPE_IP:

            # Extract destination
            string = socket.inet_ntoa(ip.dst)
            address = '.'.join(string.split(".")[:2]) #remove the last two components
            if address in subnets: #increase count in dict
                subnets[address] = subnets[address] + 1
            else: #insert key, value in dict
                subnets[address] = 1            

            # TCP packets
            if ip.p==dpkt.ip.IP_PROTO_TCP: #ip.p == 6: 
                tcpcounter+=1
                tcp=ip.data

                # HTTP uses port 80
                if tcp.dport == 80 or tcp.sport == 80:
                    httpcounter+=1

                # HTTPS uses port 443
                elif tcp.dport == 443 or tcp.sport == 443:
                    httpscounter+=1

                # SSH uses port 22
                elif tcp.dport == 22 or tcp.sport == 22:
                    sshcounter+=1

                # SMTP uses port 25
                elif tcp.dport == 25 or tcp.sport == 25:
                    smtpcounter+=1

                # telnet uses port 23
                elif tcp.dport == 23 or tcp.sport == 23:
                    telnetcounter+=1

                # whois uses port 43
                elif tcp.dport == 43 or tcp.sport == 43:
                    whoiscounter+=1

                # rsync uses port 873
                elif tcp.dport == 873 or tcp.sport == 873:
                    rsynccounter+=1

                # FTP uses port 21
                elif tcp.dport == 21 or tcp.sport == 21:
                        ftpcounter+=1

            # UDP packets
            elif ip.p==dpkt.ip.IP_PROTO_UDP: #ip.p==17:
                udpcounter+=1
                udp=ip.data

                # DHCP uses ports 67, 68
                if udp.dport == 67 or udp.dport == 68:
                    dhcpcounter+=1

                # NTP uses port 123
                elif udp.dport == 123:
                    ntpcounter+=1
                
    print ("\t\tTotal number of TCP packets :", tcpcounter)
    print ("\t\t\tTotal number of IPV6 packets :", ipv6counter)
    print ("\t\t\tTotal number of rsync packets :", rsynccounter)
    print ("\t\tTotal number of UDP packets :", udpcounter)
    print ("\t\tTotal number of ARP packets :", arpcounter)
```
