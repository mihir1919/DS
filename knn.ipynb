```python
def dis(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knnalgo(X,Y,query,k=5):
    
    val=[]
    l=[]
    for i in range(len(X)):
        
        d=dis(X[i],query)
        l.append([d,Y[i]])
    l.sort()
    
    m=0
    l=np.array(l)
    
    dic={}
    n=0
#     print(l)
    for i in l[:k]:
#         print(i[1])
        if(i[1] in dic.keys()):
            dic[i[1]]+=1

        else:
            dic[i[1]]=1
            
        if(dic[i[1]]>m):
            n=i[1]
            m=dic[i[1]]
            
    return n
        
```
