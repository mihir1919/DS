```python
l=[]

def dis(x,y):
    return (np.sqrt(sum((x-y)**2)))

def knn(xval,yval,x,y):
#     xval=xval.values
#     yval=yval.values
    
    for i in range(len(yval)):
#         print(i)
#         print(xval[i][0],xval[i][1])
        d=dis(xval[i],[x,y])
        l.append((d,yval[i]))
#         print(dis)
#     print(l)
```
