```python
import dpkt
import os
import socket
import string
import sys

# Main program
if __name__ == "__main__":
    
    tcpcounter=0.0
    udpcounter=0.0
    arpcounter=0.0
    rsynccounter=0.0
    ipv6counter=0.0

    # Subnet Dictionary
    subnets = {}

    # Open file
    print("18bci0190")
    filename="w1.pcap"
    if os.path.isfile(filename):
        print ("Present: ",filename)
        trace=filename
    else:
        print ("Absent: ",filename)
        sys.stderr.write("Cannot open file for reading\n")
        sys.exit(-1)

    # Packet processing loop
    for ts,pkt in dpkt.pcap.Reader(open(filename,'rb')):
        counter+=1
        
        # Parse ethernet packet
        eth=dpkt.ethernet.Ethernet(pkt)
        ip=eth.data       
                
        #check if IP packet or non-ip packet
        if eth.type == dpkt.ethernet.ETH_TYPE_IP or eth.type == dpkt.ethernet.ETH_TYPE_IP6:
            ipcounter = ipcounter + 1
        else:
            nonipcounter = nonipcounter + 1    

        # IPV6 packets
        if eth.type==dpkt.ethernet.ETH_TYPE_IP6: 
            ipv6counter+=1     

        # ARP packet
        elif eth.type==dpkt.ethernet.ETH_TYPE_ARP:
            arpcounter+=1

            
        # IPV4 packets
        elif eth.type==dpkt.ethernet.ETH_TYPE_IP:

            # Extract destination
            string = socket.inet_ntoa(ip.dst)
            address = '.'.join(string.split(".")[:2]) #remove the last two components
            if address in subnets: #increase count in dict
                subnets[address] = subnets[address] + 1
            else: #insert key, value in dict
                subnets[address] = 1            

            # TCP packets
            if ip.p==dpkt.ip.IP_PROTO_TCP: #ip.p == 6: 
                tcpcounter+=1
                tcp=ip.data

       
                # rsync uses port 873
                elif tcp.dport == 873 or tcp.sport == 873:
                    rsynccounter+=1

            # UDP packets
            elif ip.p==dpkt.ip.IP_PROTO_UDP: #ip.p==17:
                udpcounter+=1
                udp=ip.data

    print ("\tTotal number of IP packets :", ipcounter)
    print ("\tTotal number of IPV6 packets :", ipv6counter)
    print ("\tTotal number of IPV4 packets :", 0)
    print ("\tTotal number of TCP packets :", tcpcounter)
    print ("\tTotal number of rsync packets :", rsynccounter)

    print ("\tTotal number of UDP packets :", udpcounter)
    print ("\tTotal number of ARP packets :", arpcounter)
    print ("--------------------------------------------------------------")
```
