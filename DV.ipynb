```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(10)
y=np.sqrt((x**2)+(x**3))
# print(x)
plt.style.use("seaborn")
plt.plot(x,color="red",label="mihir",marker="o",linestyle="dashed")
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.scatter(x,y)
```

```python
x_coordinates=np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,40],width=0.5)
plt.bar(x_coordinates+0.5,[12,40,50],width=0.5)
plt.show()
```

```python
#standard deviation
Xsn = np.random.randn(100)# Random points in standard normal
# Xsn = (X - u)/sigma
sigma = 8
u = 70
X = np.round(Xsn*sigma + u)
# X2 = np.round
print(X)
```

```python
plt.style.use("seaborn")
plt.hist(X,width=0.5)
```

```python
#movie/books dv
import pandas as pd
df=pd.read_csv('books.csv')
x=list(df.get("book_title"))
x=[len(i) for i in x]
freq={}
for i in x:
    if i not in freq.keys():
        freq[i]=1
    else:
        freq[i]+=1
X=np.array(list(freq.keys()))
Y=np.array(list(freq.values()))
plt.scatter(X,Y)
```
